```python
#importing the packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np

sb.set()
```

```python
#reading the dataset
analysis = pd.read_csv('/kaggle/input/matches/dataset.csv')
```

```python
analysis
```

```python
#dropping the unnecessary columns from the dataset
analysis = analysis.drop(['umpire1'], axis=1)
analysis = analysis.drop(['umpire2'], axis=1)
analysis = analysis.drop(['umpire3'], axis=1)
analysis = analysis.drop(['dl_applied'], axis=1)
```

```python
#This code will give you total number of null values in each features in the data frame.
analysis.isnull().sum()
```

```python
#Dropping the null values in each column
analysis = analysis.dropna(axis=0)
```

```python
#This code will give you total number of null values in each features in the data frame.
analysis.isnull().sum()
```

```python
analysis
```

```python
#This code will return the details of the players with POTM awards in descending order.
analysis['player_of_match'].value_counts()[:10]
```

```python
#Data Visualisation with barplot. This code visualises who won the maximum POTM awards from the dataset.
sb.barplot(x = analysis['player_of_match'].value_counts()[:10].index, y = analysis['player_of_match'].value_counts()[:10])
plt.xticks(rotation = 60)
```

```python
#This code will return the number of unique values from player_of_match column.
analysis['player_of_match'].nunique()
```

```python
#Data Visualisation with heatmap
sb.heatmap(analysis.corr(), annot=True)
```

```python
#This code will return the count of toss won by each team.
analysis['toss_winner'].value_counts()
```

```python
#Data Visualisation with lineplot. This code visualises which team has highest toss winning streak from the dataset.
plt.figure(figsize = (10,7))
sb.lineplot(x = analysis['toss_winner'].value_counts().index, y = analysis['toss_winner'].value_counts())
plt.xticks(rotation = 90)
plt.show()
```

```python
#This code will return the number of matches won by each team.
analysis['winner'].value_counts()
```

```python
#Data Visualisation with scatterplot. This code visualises which team has highest winning streak from the dataset.
plt.figure(figsize = (15,7))
sb.scatterplot(x = analysis['winner'].value_counts().index, y = analysis['winner'].value_counts())
plt.xticks(rotation = 90)
plt.show()
```
